- 這裏是一個建立`正弦波`圖表的範例

```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸的數據點
x = np.linspace(0, 2*np.pi, 100)

# 計算正弦值
y = np.sin(x)

# 繪製圖表
plt.plot(x, y)
plt.xlabel("x")  # 設定 x 軸標籤
plt.ylabel("sin(x)")  # 設定 y 軸標籤
plt.title("sin plot")  # 設定圖表標題
plt.grid(True)  # 顯示格線
plt.show()  # 顯示圖表
```

- 餘弦波圖表

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.cos(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.title("Cosine Wave")
plt.grid(True)
plt.show()
```

- 同時顯示sin,cos的正弦波和餘弦波的圖表

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成數據
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# 繪製圖表
plt.figure(figsize=(10, 5))
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.title('sin and cos')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
